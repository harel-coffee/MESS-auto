```python
strength = 1.
trait_value = -5.32923208
filt_opt = -6.59174568179
def plot_traits(trait_value=trait_value):
    def filt(strength = strength, victim_trait=trait_value, filt_opt=filt_opt):
        return 1 - (np.exp(-(strength)*((victim_trait - filt_opt) ** 2)))
    def comp(strength = strength, victim_trait=trait_value, comp_mean=filt_opt):
        return np.exp(-((victim_trait - comp_mean) ** 2)*strength)
    fig, ax = plt.subplots(figsize=(14, 5))
    xs = np.linspace(0.0001, 20, 100)
    ax.plot(xs, [filt(x, trait_value, filt_opt) for x in xs], label="Filtering")
    ax.plot(xs, [comp(x, trait_value, filt_opt) for x in xs], label="Competitiong")        
    ax.set_ylabel("delta_i", fontsize=20)
    ax.set_xlabel("e_S", fontsize=20)
    ax.set_title("Filtering/Competition", fontsize=25)
    plt.legend(title="Assembly Model: {}".format(filt_opt-trait_value), fontsize=15)
#plot_traits()
for trait in [-6.5, -7, -7.5]:
    plot_traits(trait_value=trait)
```
