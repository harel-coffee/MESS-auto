## CV classification by hand
You can also perform the classification cv by hand, but this requires a few more
steps, including creating a dummy empirical dataset (synthetic community), and calling
predict and score by hand. This is useful for testing.

```python
## Ensemble class needs to be fed a df for the empirical data
## which is ignored by the cv routines, so here we just give
## it some dummy data.
tmp_com = MESS.util.synthetic_community()
cla = MESS.inference.Classifier(tmp_com, simfile, algorithm="rf", verbose=True)
cla.cross_val_predict(quick=True, verbose=True)
cla.cross_val_score(quick=True, verbose=True)
_ = cla.plot_confusion_matrix()
```

```python
fig4_dir = "/home/iovercast/Continuosity/MESS/analysis/figure4-sims/"
if not os.path.exists(fig4_dir):
    os.mkdir(fig4_dir)

## The full list of data combinations: MESS.stats.feature_sets().keys()
data_combos = ['abundance', 'pi', 'abundance+pi', 'pi+trait', 'abundance+pi+trait']

classifiers = {x:{} for x in data_combos}
for dstring in data_combos:
    try:
        data_combo = dstring.split("+")
    except:
            data_combo = dstring
    print("Using data axes {}".format(data_combo))

    ## Partition the simulations into different groups based on J
    j_brackets = np.linspace(0, 10000, 11)
    for i in range(len(j_brackets)-1):
        !date
        j_lo = int(j_brackets[i])
        j_hi = int(j_brackets[i+1])

        ## Grab a chunk of simulations bracketed by low/hi J values
        tmp_df = sim_df[(sim_df["J"] >= j_lo) & (sim_df["J"] <= j_hi)]
        print(j_lo, j_hi, len(tmp_df))

        ## Write the sims to a file so the cv procedure can read them in
        outfile = fig4_dir+"SIMOUT-{}-{}.txt".format(j_lo, j_hi)
        tmp_df.to_csv(path_or_buf=outfile, sep="\t", index=False)

        ## perform the CV and save the Classifier object so we can pull from it later
        classifiers[dstring][j_lo] = MESS.inference.classification_cv(simfile=outfile, quick=False,\
                                                          data_axes=data_combo,
                                                          verbose=False)
```

```python
models = ["neutral", "filtering", "competition"]

def get_cv_result(dic, model, stat):
    ''' A helper function for pulling stat values for specific models out
    of the list of classifier structure that we have. Give it the dictionary
    for a given data combo, the assembly model, and the report statistic and
    it'll return the list of statistics in order of J size of sim data. Options
    for stat are 'precision', 'recall', 'f1-score', & 'support'.
    '''
    xs = sorted(dic.keys())
    ys = []
    for x in xs:
        tmp_cla = dic[x]
        cv_rep = metrics.classification_report(y_true=tmp_cla.y, y_pred=tmp_cla.cv_preds, output_dict=True)
        ys.append(cv_rep[model][stat])
    return np.array(ys)


def plot_misclassification(class_dict, ax='', legend=False, ylabel=True, yticks=True, lw=3, verbose=False):
    if not ax:
        fig, ax = plt.subplots(figsize=(5, 5))
    xs = sorted(class_dict.keys())
    if verbose: print(xs)

    for m in models:
        ys = get_cv_result(class_dict, m, 'precision')
        ax.plot(xs, 1-ys, c=MESS.plotting.model_colors[m], label=m, lw=lw)
        ys = get_cv_result(class_dict, m, 'recall')
        #ax.plot(xs, 1-ys, c=MESS.plotting.model_colors[m], label=m + " recall", ls='--', lw=lw)
        ax.plot(xs, 1-ys, c=MESS.plotting.model_colors[m], ls='--', lw=lw)

    ## xticks
    ax.xaxis.set_ticklabels([0, 500, 2000, 4000, 6000, 8000])
    plt.setp(ax.get_xticklabels(), rotation=30, ha="right",\
            rotation_mode="anchor", fontsize=20)

    ## yticks
    ax.set_ylim(0, .75)
    plt.setp(ax.get_yticklabels(), fontsize=20)


    if not yticks: ax.yaxis.set_ticks([])
    if ylabel: ax.set_ylabel('Misclassification Rate', fontsize=25)

    if legend: ax.legend(fontsize=20)

plot_misclassification(classifiers["abundance+pi+trait"], yticks=True, legend=True)
```

```python
## Confusion matrix code stolen directly from MESS.inference, but I had to hack it a little
## to make the figure look nice and didn't feel like hacking in the changes to the MESS code
def plot_confusion_matrix(mess_cla, ax='', figsize=(8, 8), title="", normalize=True, outfile='',
                          ylabel=True, xlabel=True, yticks=True, xticks=True):
    self = mess_cla
    
    if not ax:
        fig, ax = plt.subplots(figsize=(8,8))
    labels = ["Competition", "Filtering", "Neutral"]

    conf_matrix = metrics.confusion_matrix(self.y, self.cv_preds)

    if normalize:
        conf_matrix = conf_matrix.astype(float)/conf_matrix.sum(axis=1)

    im = ax.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Greys)

    ax.set_title(title, fontsize=25)

    if xticks:
        ax.set_xticks(np.arange(conf_matrix.shape[1]))
        ax.set_xticklabels(labels, fontsize=15)
        plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
                rotation_mode="anchor")
        if xlabel: ax.set_xlabel('Predicted assembly model class', fontsize=25)
    else:
        ax.xaxis.set_ticks([])

    if yticks:
        ax.set_yticks(np.arange(conf_matrix.shape[0]))
        ax.set_yticklabels(labels, fontsize=15)
        if ylabel: ax.set_ylabel('True assembly model class', fontsize=25)
    else:
        ax.yaxis.set_ticks([])

    fmt = '.2f' if normalize else 'd'
    thresh = conf_matrix.max() / 2.
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            ax.text(j, i, format(conf_matrix[i, j], fmt),
                    ha="center", va="center", fontsize=15,
                    color="white" if conf_matrix[i, j] > thresh else "black")

plot_confusion_matrix(classifiers[d][9000], xticks=False, ylabel=True, yticks=True)
```
